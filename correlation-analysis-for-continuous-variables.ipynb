# The Pearson Correlation

```python
sb.pairplot(mr)
```

```python
corr = mr.corr()

# Generate a mask for upper traingle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Configure a custom diverging colormap
cmap = sb.diverging_palette(220, 20, as_cmap=True, center='light')

# Draw the heatmap
heatmap = sb.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot=True, cmap=cmap, mask=mask)
heatmap.set_title('Correlation Between Continuous Numerical Variables')
```
