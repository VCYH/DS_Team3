```python
import pandas as pd
import numpy as np
import os 
from sklearn.feature_selection import chi2
from scipy import stats
import seaborn as sns
import matplotlib.pylab as plt
```

```python
# Loading file
mr_df = pd.read_csv("../input/cleansedtraining-20apr-1010/Cleansed-training 20apr 1010.csv")
```

```python
mr_df.drop(['age', 'avg_glucose_level', 'bmi'], inplace=True, axis=1)
```

```python
# Extracting column names 
column_names=mr_df.columns
# Assiging column names to row indexs 
chisqmatrix=pd.DataFrame(mr_df,columns=column_names,index=column_names)
```

```python
# Setting counters to zero
outercnt=0
innercnt=0

for icol in column_names:
    
    for jcol in column_names:
        
       mycrosstab=pd.crosstab(mr_df[icol],mr_df[jcol])
       #print (mycrosstab)
       stat,p,dof,expected=stats.chi2_contingency(mycrosstab)
       chisqmatrix.iloc[outercnt,innercnt]=round(p,4)
       cntexpected=expected[expected<5].size
       perexpected=((expected.size-cntexpected)/expected.size)*100
      
       #print (icol)
       #print (jcol)
       if perexpected<20:
            chisqmatrix.iloc[outercnt,innercnt]=2
       #print (perexpected) 
       if icol==jcol:
           chisqmatrix.iloc[outercnt,innercnt]=0.00
       #print (expected) 
       innercnt=innercnt+1
    #print (outercnt) 
    outercnt=outercnt+1
    innercnt=0
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,12))         # Sample figsize in inches
sns.heatmap(chisqmatrix.astype(np.float64), annot=True, linewidths=0.5, ax=ax,
            cmap='coolwarm')
```

Any categorical pair that is less than our chosen significance level of (Î± = 0.05), we can reject the null hypothesis. We can conclude that there is enough evidence to suggest an association between them.
